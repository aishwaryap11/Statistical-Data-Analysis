### An independent two-sample z-test to examine if there is a significant difference in the mean of student grades between two groups ("Parents living together" and "Parents living apart")

```python
from statsmodels.stats.weightstats import ztest

group1 = df[df['Pstatus'] == 'T']['G3']  # Parents living together
group2 = df[df['Pstatus'] == 'A']['G3']  # Parents living apart

statistic, p_value = ztest(group1, x2=group2, value=0)

mean_group1 = np.mean(group1)
mean_group2 = np.mean(group2)
std_dev_group1 = np.std(group1)
std_dev_group2 = np.std(group2)
# Calculate the confidence interval
ci_low, ci_high = (mean_group1 - mean_group2) - 1.96 * np.sqrt(std_dev_group1**2/len(group1) + std_dev_group2**2/len(group2)), \
                  (mean_group1 - mean_group2) + 1.96 * np.sqrt(std_dev_group1**2/len(group1) + std_dev_group2**2/len(group2))

# Print results
print(f"P-value for Family Support: {p_value}")
print(f"Confidence Interval: ({ci_low:.2f}, {ci_high:.2f})")
print(f"Mean (Family Support: yes): {mean_group1:.2f}")
print(f"Mean (Family Support: no): {mean_group2:.2f}")
print(f"Standard Deviation (Family Support: yes): {std_dev_group1:.2f}")
print(f"Standard Deviation (Family Support: no): {std_dev_group2:.2f}")

```

### An independent two-sample z-test to examine if there is a significant difference in the mean of student grades between two groups ("With family support" and "Without family support")

```python
# Import necessary libraries
from statsmodels.stats.weightstats import ztest

# Assume df is your DataFrame with the relevant data
group1 = df[df['famsup'] == 'yes']['G3']  # Family support: yes
group2 = df[df['famsup'] == 'no']['G3']   # Family support: no

# Perform the two-tailed z-test
statistic, p_value = ztest(group1, x2=group2, value=0)

mean_group1 = np.mean(group1)
mean_group2 = np.mean(group2)
std_dev_group1 = np.std(group1)
std_dev_group2 = np.std(group2)
# Calculate the confidence interval
ci_low, ci_high = (mean_group1 - mean_group2) - 1.96 * np.sqrt(std_dev_group1**2/len(group1) + std_dev_group2**2/len(group2)), \
                  (mean_group1 - mean_group2) + 1.96 * np.sqrt(std_dev_group1**2/len(group1) + std_dev_group2**2/len(group2))

# Print results
print(f"P-value for Family Support: {p_value}")
print(f"Confidence Interval: ({ci_low:.2f}, {ci_high:.2f})")
print(f"Mean (Family Support: yes): {mean_group1:.2f}")
print(f"Mean (Family Support: no): {mean_group2:.2f}")
print(f"Standard Deviation (Family Support: yes): {std_dev_group1:.2f}")
print(f"Standard Deviation (Family Support: no): {std_dev_group2:.2f}")

```

#### Same hypothesis with chi-sq test

```python
from scipy.stats import chi2_contingency

# Assume df is your DataFrame with the relevant data
contingency_table = pd.crosstab(df['Pstatus'], df['G3'])

# Perform the chi-square test
statistic, p_value, _, _ = chi2_contingency(contingency_table)

# Check the p-value
print(f"P-value: {p_value}")

# Compare the p-value to the significance level (e.g., 0.05) to make a decision
```
